```python
import os
import asyncio
import aiofiles
import numpy as np
import pydicom
import ipywidgets as widgets
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import ipyvolume as ipv
import SimpleITK as sitk
import torch
import torchvision.transforms as T
import torchio as tio
from torch.multiprocessing import Pool, cpu_count
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
import nest_asyncio
nest_asyncio.apply()


# Setting device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Loading the Data
# Copy files from Google Drive to local Colab storage
# !cp -r "/content/drive/MyDrive/3D-3D_Image_Registration/3D-3D_Image_Registration/SE000003" "/content/local_data/"

# Now change your data_path to point to the local storage
data_path = r"C:\Users\HP\Documents\GitHub\3D-3D_Image_Registration\SE000003"

# List all files in the directory
files = os.listdir(data_path)
print(files)




### We define the segmentation functions
# TorchIO expects PyTorch tensors as inputs, so the segmentation functions must accept and return PyTorch tensors.
def segment_lung(image_tensor):
    # Apply a threshold to separate lung pixels from others
    lung_mask = (image_tensor > -1000) & (image_tensor < -400)

    return lung_mask.float().to(device)


def segment_bones(image_tensor):
    # Use thresholding to identify the bones
    bone_mask = image_tensor > 300

    return bone_mask.float().to(device)

# Your segmentation functions and other functions remain unchanged

async def read_file_as_byte(path):
    async with aiofiles.open(path, 'rb') as f:
        return await f.read()

async def load_dicom_series_async(directory_path):
    # List all files in the directory
    files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
    tasks = [read_file_as_byte(os.path.join(directory_path, f)) for f in files]
    slices = await asyncio.gather(*tasks)
    slices = [pydicom.read_file(pydicom.filebase.DicomBytesIO(s)) for s in slices]
    # Convert the pixel data to float32
    slices = [torch.from_numpy(s.pixel_array.astype(np.float32)) for s in slices]
    # Concatenate the slices into a 3D tensor and move to GPU
    return torch.stack(slices).to(device)

# Use asyncio.run to run the main function
ct_image = asyncio.run(load_dicom_series_async(data_path))


### Loading the DICOM images:
# # Define a helper function to read a DICOM file into a PyTorch tensor
# def load_dicom_file(path):
#     # Load the DICOM file
#     dicom_file = pydicom.dcmread(path)

#     # Convert the pixel data to float32
#     pixel_data = dicom_file.pixel_array.astype(np.float32)

#     # Convert to a PyTorch tensor and return
#     return torch.from_numpy(pixel_data)

# def load_dicom_series(directory_path):
#     # List all files in the directory
#     files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]

#     # Use a multiprocessing pool to read the files in parallel
#     with Pool(cpu_count()) as p:
#         slices = p.map(load_dicom_file, [os.path.join(directory_path, f) for f in files])

#     # Concatenate the slices into a 3D tensor and move to GPU
#     return torch.stack(slices).to(device)




###Then, we will display the images:
def display_images(image_tensor, title, montage_slices=10):
    # Normalize to 0-255
    image_tensor = ((image_tensor - image_tensor.min()) * (255 / (image_tensor.max() - image_tensor.min()))).to('cpu').numpy().astype('uint8')

    # 3D visualization using ipyvolume
    ipv.figure()
    ipv.volshow(image_tensor, level=[0.25, 0.75], opacity=0.03, level_width=0.1, data_min=0, data_max=255)
    fig3d = ipv.gcc()

    # Create a layout with a title
    layout = widgets.VBox([widgets.HTML(f'<h2>{title} - 3D View</h2>'), fig3d])
    display(layout)

    # Montage of slices using matplotlib
    fig = plt.figure(figsize=(10, 2))
    fig.suptitle(title + " - Slice View")
    slice_interval = image_tensor.shape[0] // montage_slices
    for i in range(montage_slices):
        ax = fig.add_subplot(1, montage_slices, i + 1)
        ax.imshow(image_tensor[i * slice_interval], cmap='gray')
        plt.axis('off')
    plt.show()



###The get_initial_transform function should be defined to provide an initial estimate of the transformation needed to register the moving image to the fixed image.
##This can be calculated in many ways depending on the specifics of your images and the registration task. Here is an example of how we can define this function,
### assuming that we are performing rigid registration:
def get_initial_transform(fixed_image, moving_image):
    # Initialize the transform
    initial_transform = sitk.CenteredTransformInitializer(sitk.Cast(fixed_image,sitk.sitkFloat32),
                                                          sitk.Cast(moving_image,sitk.sitkFloat32),
                                                          sitk.Euler3DTransform(),
                                                          sitk.CenteredTransformInitializerFilter.GEOMETRY)

    return initial_transform



###Then, we load the DICOM images, perform lung and bone segmentation, and display the results:
# Load the DICOM images
# ct_image = load_dicom_series(data_path)

# Perform lung segmentation
lung_mask = segment_lung(ct_image)

# Display segmented lungs in 3D
display_images(lung_mask, "Segmented Lungs")

# Perform bone segmentation
bone_mask = segment_bones(ct_image)

# Display the bone segmentation in 3D and slices
display_images(bone_mask, "Segmented Bones")



###For the transformation part, I'll leave it as is because it's not parallelizable and does not benefit from being on the GPU.
# Paths to the DICOM directories
fixed_image_path = "/content/local_data/"
moving_image_path = "/content/local_data/"

# Load the DICOM images
fixed_image = load_dicom_series_async(fixed_image_path)
moving_image = load_dicom_series_async(moving_image_path)

# Display the images with appropriate titles
display_images(fixed_image, "Fixed Image patient's CT scan")
display_images(moving_image, "Moving Image CT scan of the phantom")

# Get the initial transform
initial_transform = get_initial_transform(fixed_image, moving_image)

# Apply the transformation
registration_method = sitk.ImageRegistrationMethod()
registration_method.SetMetricAsMeanSquares()
registration_method.SetOptimizerAsGradientDescent(learningRate=1.0, numberOfIterations=100, convergenceMinimumValue=1e-6, convergenceWindowSize=10)
registration_method.SetInterpolator(sitk.sitkLinear)
registration_method.SetInitialTransform(initial_transform, inPlace=False)
final_transform_v1 = registration_method.Execute(sitk.GetImageFromArray(fixed_image.cpu().numpy()),
                                                 sitk.GetImageFromArray(moving_image.cpu().numpy()))

# Resample the moving image onto the fixed image's grid
resampler = sitk.ResampleImageFilter()
resampler.SetReferenceImage(sitk.GetImageFromArray(fixed_image.cpu().numpy()))
resampler.SetTransform(final_transform_v1)
resampled_moving_image = torch.from_numpy(sitk.GetArrayFromImage(resampler.Execute(sitk.GetImageFromArray(moving_image.cpu().numpy())))).to(device)

# Display the images after transformation
display_images(fixed_image, "Fixed Image after Transformation")
display_images(resampled_moving_image, "Resampled Moving Image after Transformation")
```
